###`Problem 1:` Combine two lists index-wise(columns wise)

Write a program to add two lists index-wise. Create a new list that contains the 0th index item from both the list, then the 1st index item, and so on till the last element. any leftover items will get added at the end of the new list.

`Given List:`
```
list1 = ["M", "na", "i", "Sho"]
list2 = ["y", "me", "s", "vo"]
```

`Output:`
```
[['M','y'], ['na', me'], ['i', 's'], ['Sho', 'vo']]
```

````python
# Write code here
list1 = ["M", "na", "i", "Sho"]
list2 = ["y", "me", "s", "vo"]

result = [[a , b] for a, b in zip(list1, list2)]
result
````

###`Problem 7:` Sort a list of alphanumeric strings based on product value of numeric character in it. If in any string there is no numeric character take it's product value as 1.

`Input:`
```
['1ac21', '23fg', '456', '098d','1','kls']
```

`Output:`
```
['456', '23fg', '1ac21', '1', 'kls', '098d']

```

````python
# Write code here
lst = ['1ac21', '23fg', '456', '098d', '1', 'kls']

def digit_product(s):
    digits = [int(ch) for ch in s if ch.isdigit()]

    if not digits:
        return 1

    product = 1
    for d in digits:
        product *= d

    return product

result = sorted(lst, key=digit_product, reverse=True)

print(result)
````

###`Q16:` Join Tuples if similar initial element
While working with Python tuples, we can have a problem in which we need to perform concatenation of records from the similarity of initial element. This problem can have applications in data domains such as Data Science.

For eg.
```
Input  : test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
Output : [(5, 6, 7, 8), (6, 10), (7, 13)]
```

````python
# write your code here
list9 = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
result = {}
for a, b in list9:
    if a in result.keys():
        result[a].append(b)
    else:
        result[a] = [a, b]
[tuple(v) for v in result.values()]
````

###`Q17:` Multiply Adjacent elements (both side) and take sum of right and lest side multiplication result.


For eg.
```
The original tuple : (1, 5, 7, 8, 10)
Resultant tuple after multiplication :

(1*5, 1*5+5*7, 7*5 + 7*8, 8*7 + 8*10, 10*8) -> (5, 40, 91, 136, 80)

output-(5, 40, 91, 136, 80)
```

````python
# write your code here
t = (1, 5, 7, 8, 10)
list10 = []
for i in range(len(t)):
    if i==0:
        list10.append(t[i]*t[i+1])
    elif i==len(t)-1:
        list10.append(t[i]*t[i-1])
    else:
        list10.append(t[i-1]*t[i]+t[i]*t[i+1])
list10
````

###`Q19`: Count no of tuples, list and set from a list
```
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

```
`Output:`

```
List-2
Set-2
Tuples-1
```

````python
# write your code here
list1 = [{'hi', 'bye'}, {'Geeks', 'forGeeks'}, ('a', 'b'), ['hi', 'bye'], ['a', 'b']]

count_list = 0
count_set = 0
count_tuple = 0

for item in list1:
    if isinstance(item, list):
        count_list += 1
    elif isinstance(item, set):
        count_set += 1
    elif isinstance(item, tuple):
        count_tuple += 1

print("List-", count_list)
print("Set-", count_set)
print("Tuples-", count_tuple)
````
