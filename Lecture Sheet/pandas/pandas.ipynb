```python
#next function is isin()

# Q. Find the most destructive death over batsman in the history of IPL i.e who has scored the most runs in overs 16-20

# strike_rate = (total runs/ total balls faced) * 100
# Min balls faced to be considered = 200 balls in overs (16-20 overs = 5 overs = 30 balls)

# steps:
# 1. filter data for overs 16-20
# 2. group by batsman and aggregate runs and balls faced
# 3. filter batsman with min 200 balls faced
# 4. calculate strike rate
# 5. sort by strike rate and fetch top 5

```

```python
over_mask = delivary['over']>15 
death_overs = delivary[over_mask]
death_overs.head()
```

```python
all_batsman=death_overs.groupby('batsman')['batsman_runs'].count()
all_batsman
```

```python
x = all_batsman>200
qualified_batsman = all_batsman[x]

batsman_list = qualified_batsman.index.tolist()
```

```python
# Runs scored by qualified batsman in death overs
delivary['batsman'].isin(batsman_list)
final = delivary[delivary['batsman'].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
runs.sample(5)
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
balls.sample(5)
```

```python
sr = (runs/balls)*100
sr.sample(5)
```
