```python
# mask where batsman_run ==4
mask = delivary['batsman_runs'] == 4
four_runs = delivary[mask]
four_runs.shape
```

```python
four_runs['batsman'].value_counts().sort_values(ascending=False).head()
```

```python
six_runs_mask = delivary['batsman_runs'] == 6
six = delivary[six_runs_mask]
six
```

```python
six['batsman'].value_counts().sort_values(ascending=False).head()
```

```python
six.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)
```

```python
# 3 highest runs batted by a batsman against 3 different teams

batsman_team = delivary.groupby(['batsman', 'bowling_team'])
ind=batsman_team['batsman_runs'].sum().sort_values(ascending=False)
```

```python
def run_scrore(batsman_name):
    return ind[batsman_name].head(3)

run_scrore('MS Dhoni')
```
